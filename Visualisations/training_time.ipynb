```python
import matplotlib.pyplot as plt
import numpy as np
import time
import seaborn as sns
```

```python
# Average training times and RMSE values for each model
models = ['CNN-LSTM', 'CNN-LSTM Improved', 'Transformer-LSTM', 'XGBoost-LSTM']
training_times = [49.1, 122.8, 94.63, 119.4]  # in seconds
rmse_values = [18.96, 9.71, 18.00, 24.49]

# bar chart for training time
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
ax = sns.barplot(y=models, x=training_times, palette="viridis")

# Add the values at the end of each bar
for i, v in enumerate(training_times):
    ax.text(v + 1, i, f"{v:.1f}s", va='center', fontweight='bold')

# Extend the x-axis limit to accommodate the text
plt.xlim(0, max(training_times) * 1.1)

plt.xlabel('Training Time (seconds)')
plt.title('Model Training Time Comparison')
plt.tight_layout()
plt.savefig('training_time_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# 2. Time vs Performance scatterplot
plt.figure(figsize=(10, 8))
sns.set_style("whitegrid")

# Diff colors for diff models
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B3']  
scatter = plt.scatter(training_times, rmse_values, s=100, 
                     alpha=0.7, color=colors)  

# Add model names as annotations
for i, model in enumerate(models):
    plt.annotate(model, (training_times[i], rmse_values[i]), 
                 xytext=(0, 10), textcoords='offset points', 
                 ha='center', va='bottom')

plt.xlabel('Training Time (seconds)')
plt.ylabel('RMSE (lower is better)')
plt.title('Performance vs. Training Time Trade-off')
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('performance_vs_time.png', dpi=300, bbox_inches='tight')
plt.show()
```
