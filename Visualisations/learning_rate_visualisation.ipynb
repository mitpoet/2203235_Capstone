# Learning Rate Visualization

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

# Load all three history files
with open('cnn-lstm-best.pkl', 'rb') as file:
    cnn_lstm_history = pickle.load(file)
    
with open('model_history.pkl', 'rb') as file:
    transformer_lstm_history = pickle.load(file)
    
with open('xgboost-lstmBest.pkl', 'rb') as file:
    xgboost_history = pickle.load(file)


plt.figure(figsize=(12, 6))

# Plot learning rate for each model
plt.plot(cnn_lstm_history['learning_rate'], 
         label='CNN-LSTM', marker='o', markersize=4)

plt.plot(transformer_lstm_history['learning_rates'], 
         label='Transformer-LSTM', marker='s', markersize=4)

plt.plot(xgboost_history['learning_rate'], 
         label='XGBoost', marker='^', markersize=4)

# Labels and formatting
plt.title('Learning Rate Comparison Across Models', fontsize=16)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Learning Rate', fontsize=14)
plt.yscale('log') 
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('learning_rate_comparison.png', dpi=300)
plt.show()
```
